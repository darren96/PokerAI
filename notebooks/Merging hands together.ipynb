```python
def merge_dicts(dict1, dict2):
    res = dict2
    for key in dict1:
        for item in dict1[key]:
            res[key][item] = dict1[key][item]
    return res
```

```python
merge_dicts(dict1[14], dict1[13])
```

```python
def merge_results(results):
    dct = {}
    for i in range(-1,18):
        dct[i] = {}
    for result in results:
        dct = merge_dicts(result, dct)
    return dct
```

```python
res = merge_results(dict1)
```
